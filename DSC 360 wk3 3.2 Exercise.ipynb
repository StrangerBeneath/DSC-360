# <font color=2d5db5>**4A. Assuming you installed Tesseract, use pytesseract to read the Bowers text image.**

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe' #needed to run pytesseract -- https://github.com/madmaze/pytesseract
Bowers=(pytesseract.image_to_string('C:\\Users\\michaelmontana\\Documents\\000 - Bellevue University\\Classes\\DSC 360 Data Mining - Text Analytics and Unstructured Data\\Assignments\\Wk3\\bowers.jpg'))
print(Bowers)
```

# <font color=2d5db5>**4B. Format the resulting text by removing the extraneous characters.**

```python
Bowersb=Bowers
Bowersb = Bowersb.replace('\n',' ').replace('  ','').replace("‘",'').replace('”','').replace('“','')#removing line feed, double space, apostrophe, and quotation marks w/ .replace -- https://python-reference.readthedocs.io/en/latest/docs/str/replace.html
Bowersb = Bowersb.translate(str.maketrans('', '', string.punctuation))#removing punctuation using .translate -- https://python-reference.readthedocs.io/en/latest/docs/str/translate.html;  string.punctuation -- https://docs.python.org/3/library/string.html#string.punctuation
print(Bowersb)
```

# <font color=2d5db5>**4C. Finally, use spaCy to print out the tokens (the text, part of speech, and dependency). (HINT: you may need to print the tokens to see what characters you need to remove.)**

```python
nlp = spacy.load("en_core_web_sm") #pipelining spacy package -- https://spacy.io/api/top-level#spacy.load
doc = nlp(Bowersb) # creates a "container for accessing linguistic annotations" -- https://spacy.io/api/doc
print(f"{'Text':<15} {'Part-of-Speech (POS)':>40} {'Syntactic Dependency (DEP)':>50}\n") # making/formatting column headers
for token in doc: # using for loop to parse out text and tokens into columns
    if not nlp.vocab[token.text].is_stop: # excluding stop words from the output -- https://www.youtube.com/watch?v=laA6ffxUX-Q
        print(f'{token.text:<15} {spacy.explain(token.pos_):>40} {spacy.explain(token.dep_):>50}') #spacy.explain provides long-form description -- https://spacy.io/api/top-level#spacy.explain
```

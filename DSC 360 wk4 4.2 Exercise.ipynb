# <font color=2d5db5>**3. Using spaCy and NLTK, show the tokens, lemmas, parts of speech, and dependencies in the first 1,021 characters of big.txt.**

```python
text=big[:1021]
nlp = spacy.load("en_core_web_sm") #pipelining spacy package -- https://spacy.io/api/top-level#spacy.load
doc = nlp(text) # creates a "container for accessing linguistic annotations" -- https://spacy.io/api/doc
print('SpaCy Tokens:\n',f"{'Token:':<15} {'Lemma:':<20}{'Part-of-Speech (POS):':<30} {'Syntactic Dependency (DEP):':<40}") # making/formatting column headers
for token in doc: # using for loop to parse out text and tokens into columns
    print(f'{token.text:<15}{token.lemma_:<20}{spacy.explain(token.pos_):<30} {spacy.explain(token.dep_):<40}') #spacy.explain provides long-form description -- https://spacy.io/api/top-level#spacy.explain
```

```python
import nltk
from nltk.stem import WordNetLemmatizer

text=big[:1021]
wordnet_lemmatizer = WordNetLemmatizer()
text = nltk.word_tokenize(text) #tokenizing text
# Sets objects for each requirement
pos = nltk.pos_tag(text) #token and part of speech
lemma = [wordnet_lemmatizer.lemmatize(word) for word in text]
print('NLTK Tokens:\n',f"{'(Token,Part-of-Speech (POS)):'} {'Lemma:':>15}") # making/formatting column headers
for v, lem in zip(pos, lemma): #zipping together pos/lemma to use in for loop
    print(f"{v} {lem:>30}")
```
